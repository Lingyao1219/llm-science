```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from PIL import Image
from matplotlib.font_manager import FontProperties

garamond_font = FontProperties(fname='/content/EBGaramond-VariableFont_wght.ttf')  # Update the path to Garamond.ttf

### COUNTRIES
data_adjusted = {
    "Metric": [
        "Number of Nodes", "Number of Edges", "Density",
        "Average Shortest Path Length", "Clustering Coefficient", "Number of Components",
        "Size of Largest Component", "Power-law Exponent", "Number of Communities (Louvain)",
        "Number of Communities (CNM)", "Degree Centrality (top 5)", "Betweenness Centrality (top 5)",
        "Closeness Centrality (top 5)", "Eigenvector Centrality (top 5)"
    ],
    "Countries": [
        155, 1745, 0.15, 9.36, 0.77,
        16, 140, 5.22, 20, 19,
        "US (109),GB (91),CN (81),\nDE (79),CA (75)",
        "PL (0.056),GB (0.053),CA (0.051),\nFR (0.050),TR (0.050)",
        "PT (0.157),MX (0.157),QA (0.155),\nTR (0.154),SY (0.154)",
        "US (0.630),CN (0.498),GB (0.395),\nDE (0.222),CA (0.194)"
    ]
}

# Create a DataFrame with adjusted data
df_adjusted = pd.DataFrame(data_adjusted)

# Display the adjusted table next to the image
fig = plt.figure(figsize=(25, 10))
gs = GridSpec(1, 2, width_ratios=[3, 1])

# Image Plot
ax1 = fig.add_subplot(gs[0])  # Move the image to the left
image_path = "/content/countries_network.png"
image = Image.open(image_path)
ax1.imshow(image)
ax1.axis('off')

# Table Plot
ax0 = fig.add_subplot(gs[1])  # Move the table to the right
ax0.axis('tight')
ax0.axis('off')
table = ax0.table(cellText=df_adjusted.values, colLabels=df_adjusted.columns, cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(20)
table.scale(1.5, 1.5)

# Set the font properties for the table
for key, cell in table.get_celld().items():
    cell.set_text_props(fontproperties=garamond_font, fontsize=14)

for (i, j), cell in table.get_celld().items():
    if i in [11, 12, 13, 14]:
        cell.set_height(0.08)
    # Adjust column widths
    if j == 0:
        cell.set_width(0.8)
    elif j == 1:
        cell.set_width(0.8)

plt.show()
```
